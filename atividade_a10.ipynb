```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('os_worldwide.csv')
dados['Date'] = pd.to_datetime(dados['Date'])
```

```python
# Gráfico em linha 
so = input('Selecione o sistema operacional: ')

def plotar_grafico_linha(so):
    plt.figure(figsize=(10, 6))
    plt.plot(dados['Date'], dados[so], label=so, color='green')
    plt.title(f'Participação de Mercado do {so} ao Longo do Tempo')
    plt.xlabel('Data')
    plt.ylabel('Participação de Mercado (%)')
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

plotar_grafico_linha(so)
```

```python
def input_sos():
    input_usuario = input("Digite três sistemas operacionais, separados por vírgula (ex: Windows,Android,iOS): ")
    sos = input_usuario.split(',')
    
    sos = [so.strip() for so in sos]
    return sos

def plotar_grafico_area(sistemas_operacionais):
    plt.figure(figsize=(10, 6))
    cores = ['blue', 'green', 'red']  
    for i, sistema in enumerate(sistemas_operacionais):
        plt.stackplot(dados['Date'], dados[sistema], labels=[sistema], colors=[cores[i]], alpha=0.5)
    plt.title('Participação de mercado ao longo do tempo')
    plt.xlabel('Data')
    plt.ylabel('Participação de Mercado (%)')
    plt.legend(loc='upper left')
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

sistemas_operacionais = input_sos()
plotar_grafico_area(sistemas_operacionais)
```

```python
def plotar_grafico_barras(data_selecionada):
    
    dados['Date_str'] = dados['Date'].dt.strftime('%Y-%m')
    
    dados_selecionados = dados[dados['Date_str'] == data_selecionada]
    
    if dados_selecionados.empty:
        print("Não foram encontrados dados para a data selecionada.")
        return
    
    os_nomes = dados_selecionados.columns[1:-1] 
    os_valores = dados_selecionados.iloc[0, 1:-1].values
    
    # Gráfico de barras
    plt.figure(figsize=(12, 8))
    plt.barh(os_nomes, os_valores, color='skyblue')
    plt.title(f'Participação de Mercado dos Sistemas Operacionais em {data_selecionada}')
    plt.xlabel('Participação de Mercado (%)')
    plt.ylabel('Sistemas Operacionais')
    plt.tight_layout()
    plt.show()

data_selecionada = input("Digite a data no formato YYYY-MM (ex: 2023-01): ")

plotar_grafico_barras(data_selecionada)
```

```python
so = input('Selecione o sistema operacional: ')

def plotar_histograma(so):
    # Histograma 
    plt.figure(figsize=(10, 6))
    plt.hist(dados[so], bins=20, color='green', edgecolor='black')
    plt.title(f'Distribuição da Participação de Mercado do {so}')
    plt.xlabel('Participação de Mercado (%)')
    plt.ylabel('Frequência')
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()

plotar_histograma(so)
```

```python
# Gráfico de dispersão

so1 = input('Selecione o primeiro sistema operacional: ')
so2 = input('Selecione o segundo sistema operacional: ')

def plotar_dispersao(so1, so2):
    plt.figure(figsize=(10, 6))
    plt.scatter(dados[so1], dados[so2], color='purple', alpha=0.5)
    plt.title(f'Comparação da Participação de Mercado: {so1} vs. {so2}')
    plt.xlabel(f'{so1} (%)')
    plt.ylabel(f'{so2} (%)')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

plotar_dispersao(so1, so2)
```

```python
# Gráfico boxplot
so = input('Selecione o sistema operacional: ')

def plotar_boxplot(so):
    plt.figure(figsize=(8, 6))
    plt.boxplot(dados[so], vert=False, patch_artist=True, showmeans=True)
    plt.title(f'Distribuição da Participação de Mercado do {so}')
    plt.xlabel('Participação de Mercado (%)')
    plt.yticks([1], [so])
    plt.grid(True, axis='x')
    plt.tight_layout()
    plt.show()

plotar_boxplot(so)
```
