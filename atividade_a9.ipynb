**Operating System Market (2009 - 2023)**



*   A base de dados escolhida contém os dados referentes à porcentagem de mercado de cada SO de 2009 à 2023. Com ela é possível entender quais SO's eram dominantes, possíveis tendências do mercado, impacto de lançamentos e mudanças nas preferências de consumo.
*   A base é fornecida por Michal Bogacz em https://www.kaggle.com/datasets/michau96/operating-system-market-2009-2023?resource=download e a coleta dos dados foi realizada de 01/2009 à 12/2023.
*   A base contém 22 atributos que podem ser classificados como:
 *  Data: É um atributo nominal, que neste contexto é utilizado como um identificador para observações ao invés de para operações matemáticas.
 *  Windows, Android, iOS, OS X, Desconhecido, Linux, Series 40, SymbianOS, Samsung, BlackBerry OS, Chrome OS, Nokia Desconhecido, Playstation, Sony Ericsson, KaiOS, Xbox, bada, Tizen, LG, Nintendo, Outro: São atributos contínuos que representam as porcentagens de participação de mercado de diversos sistemas operacionais e são expressos como números flutuantes, adequados para operações matemáticas e análise.
 * É possível encontrar informações publicadas sobre esta base de dados em https://gs.statcounter.com/os-market-share, como estatísticas do mercado de SO's.  







```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/os_worldwide.csv')

variavel = input("Escolha uma variável para a Tabela de Distribuição de Frequência: ")

if variavel not in data.columns:
    print(f"A variável escolhida '{variavel}' não existe na base de dados, escolha outra.")
else:
    bins = [0, 20, 40, 60, 80, 100]

    frequencia = pd.cut(data[variavel], bins=bins, right=False).value_counts().sort_index()

    frequencia.plot(kind='bar')
    plt.title(f'Tabela de Distribuição de Frequência - {variavel}')
    plt.xlabel('Intervalos de Percentual')
    plt.ylabel('Frequência')
    plt.xticks(rotation=45)
    plt.grid(axis='y')
    plt.show()
```
