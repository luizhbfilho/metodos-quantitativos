##### Exercício 2.28 
**Uma droga para alívio da asma pode ser comprada de cinco diferentes indústrias, em forma líquida, tablete ou cápsula, todas com concentração regular ou extra. Em quantas maneiras diferentes um médico pode prescrever essa droga para o paciente que sofre de asma?**

O cálculo pode ser realizado utilizando o princípio fundamental da contagem, multiplicando 5 diferentes tipos de indústria, 3 formas e 2 concentrações resultando em: 5 x 3 x 2 = 30 maneiras de preescrever a droga.

##### Exercício 2.58
**Uma indústria automobilística está preocupada com um possível recall de seu sedã quatro portas mais vendido. Se houver um recall, há 0,25 de probabilidade de que o defeito seja no sistema de freios; 0,18 de que seja na transmissão; 0,17 de que seja no sistema de combustível e 0,40 de que seja em alguma outra parte.**

O problema apresentado possui as seguintes probabilidades:
- Probabilidade de defeito no sistema de freios: P(F) = 0,25
- Probabilidade de defeito na transmissão: P(T) = 0,18
- Probabilidade de defeito no sistema de combustível: P(C) = 0,17
- Probabilidade de defeito em alguma outra parte: P(O) = 0,40
- Probabilidade de defeitos em ambos os sistemas, simultaneamente (freios e combustível): P(F ∩ C) = 0,15

Para resolver os problemas, podemos utilizar a fórmula para a probabilidade de união de dois eventos: P(A ∪ B) = P(A) + P(B) − P(A ∩ B)

**(a) Qual é a probabilidade de que o defeito esteja nos freios ou no sistema de combustível, se a probabilidade de defeitos em ambos os sistemas, simultaneamente, é de 0,15?**

Utilizando a fórmula, obtemos:

P(A) é a probabilidade do evento A (defeito nos freios) = 0,25

P(B) é a probabilidade do evento B (defeito no sistema de combustível) = 0,17

P(A ∩ B) é a probabilidade de que ambos os eventos A e B ocorram ao mesmo tempo = 0,15

Calculando:

P(A ∪ B) = 0,25 + 0,17 - 0,15 = 0,27 = 27% de probabilidade de que o defeito esteja nos freios ou no sistema de combustível.

**(b) Qual é a probabilidade de que não haja defeitos nem no sistema de freios nem no sistema de combustível?**

Para calcular a probabilidade de nenhum defeito é necessário utilizar o último resultado obtido, que pode ser calculada por:

1 - P(A ∪ B) = 1 - 0,27 = 0,73 = 73% de não haver defeitos nem no sistema de freios e nem no sistema de combustível.

##### Exercício 2.75
**Como a situação do Exercício 2.73 pode sugerir, os procedimentos estatísticos são frequentemente usados para o controle de qualidade (ou seja, controle de qualidade industrial). Às vezes, o peso de um produto é uma variável importante a ser controlada. São dadas as especificações para peso de certo produto empacotado e o pacote é rejeitado se estiver muito pesado ou muito leve. Os dados históricos sugerem que 0,95 é a probabilidade de que o produto atenda às especificações de peso, enquanto 0,002 é a probabilidade de que o produto seja leve demais. Para cada pacote do produto o fabricante investe 20,00 na produção e o preço de compra pelo cliente é de 25,00.**

Os dados fornecidos são: 
- A probabilidade de que o produto atenda às especificações de peso é de 0,95.
- A probabilidade de que o produto seja leve demais é de 0,002.
- O custo de produção por pacote é de $20,00.
- O preço de venda por pacote ao cliente é de $25,00.

**(a) Qual é a probabilidade de que um pacote escolhido aleatoriamente da linha de produção esteja muito pesado?**

P(Pesado demais) = 1 - P(Atende às especificações) - P(Leve demais) = 1 − 0,95 − 0,002 = 0,048 = 4,8% de probabilidade de um pacote estar muito pesado.

**(b) Para cada 10.000 pacotes vendidos, qual é o lucro do fabricante se todos os pacotes atendem às especificações?**

Lucro total = (25 - 20) x 10000 = 5 x 10000 = $50000 

**(c) Assumindo-se que todos os pacotes ‘defeituosos’ sejam rejeitados e não rendam lucro nenhum, de quanto será o lucro reduzido em 10.000 pacotes devido a falha de não se atender às especificações?**

P(Defeituoso) = 0,002 + 0,048 = 0,05

Para calcular o número esperado de pacotes defeituosos: 10000 x 0,05 = 500

A redução do lucro seria calculada por: 500 x 5 = $2500

Portanto, a cada 10000 pacotes vendidos é esperada uma redução no lucro de $2500.
